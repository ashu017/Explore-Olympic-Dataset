```python

goldm = gold[gold.Sex=='M']
goldf = gold[gold.Sex=='F']
goldf.head()
```

```python
notnullgoldm = goldm[(goldm['Height'].notnull()) & (goldm['Weight'].notnull()) ]
weilif = notnullgoldm[(notnullgoldm.Sport=="Gymnastics") | (notnullgoldm.Sport=="Basketball")]
```

```python
plt.figure(figsize=(12, 10))
ax = sns.scatterplot(x="Height", y="Weight", data=weilif)
plt.title('Height vs Weight of Olympic Medalists')
```

> On this data we can apply K means clustering to cluster two sports. Weightlifting and basketball

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=2)
km
```

```python
ypre = km.fit_predict(weilif[['Height', 'Weight']])
ypre
```

```python
weilif['Cluster'] = ypre
w1 = weilif[weilif.Cluster==0]
w2 = weilif[weilif.Cluster==1]
plt.figure(figsize=(12, 10))
plt.scatter(w1.Height, w1.Weight , color="blue")
plt.scatter(w2.Height, w2.Weight , color="red")
plt.xlabel('Height')
plt.ylabel('Weight')
plt.legend()
```
